```python
#change numerical to one-hot

def y_to_onehot(y):
    yy = list()
    for i in range(len(y)):
        if y[i] == 0:
            yy.append([1,0,0,0])
        elif y[i] == 2:
            yy.append([0,1,0,0])
        elif y[i] == 3:
            yy.append([0,0,1,0])
        elif y[i] == 6:
            yy.append([0,0,0,1])
    return yy

yy = y_to_onehot(y)
yy
```
